## 1. Interpolation

Comment gérer un trou dans des données, ou quand on a deux capteurs ne mesurant pas à la même vitesse ?

```python
from scipy.interpolate import interp1d
```

```python
x = np.linspace(0, 10, 5)
y = x**2

plt.plot(x, y)
```

```python
# On voit qu'il nous manque de la précision, on va donc interpoler.
# Il existe plusieurs type d'interpolation :
# - linear (fait une ligne entre les points)
# - quadratic
# - cubic 

f = interp1d(x, y, kind='linear')

x_new = np.linspace(0, 10, 20)
y_new = f(x_new)

plt.scatter(x, y)
plt.scatter(x_new, y_new, c='r')
```

```python
# ATTENTION à ne pas trop interpoler et cacher une réalité :
x = np.linspace(0, 30, 10)
y = np.cos(x) + x/2

plt.plot(x, y)
```

```python
f = interp1d(x, y, kind='linear')

x_new = np.linspace(0, 30, 50)
y_new = f(x_new)

plt.scatter(x, y)
plt.scatter(x_new, y_new, c='r')
```

```python
# Alors que la réalité est plutot celle ci :
x = np.linspace(0, 30, 50)
y = np.cos(x) + x/2

plt.plot(x, y)
plt.scatter(x_new, y_new, c='r')
```

## 2. Optimize
### 2.1. Curve fit

On a un nuage de point, et on veut trouver une courbe la plus fidèle au nuage de points. Et oui, ça utilise les moindres carrés !

Créons un modèle :
$f(x) = a x^3 + b x^2 + c x + d$

```python
from scipy import optimize
```

```python
# on créé des data:
x = np.linspace(0, 2, 200)
y = (x**3)/3 - 2*(x**2)/3 + x/10 + 2 + np.random.randn(x.shape[0])/30

plt.scatter(x, y)
```

```python
# On sait que notre function est un polynome de degré 3 :
def f(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d
```

```python
# Maintenant on va tenter de trouver les paramètres :
params, covariance = optimize.curve_fit(f, x, y)
print(f"""
a :
    réel = 1/3
    trouvé = {params[0]}
    
b :
    réel = -2/3
    trouvé = {params[1]}
    
c :
    réel = 1/10
    trouvé = {params[2]}
    
d :
    réel = 2
    trouvé = {params[3]}
""")
```

```python
# Maintenant on test :
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]), c='r')
```

## 3. Minimize
Trouvons le minimum d'une fonction

```python
def f(x):
    return x**2/20 + np.sin(x)

x = np.linspace(-10, 10, 100)
plt.plot(x, f(x))
```

```python
# Il y a pleins de fonctions de minimization (c.f. la doc).
# Cependant, certaines tolèrent bien les minima locaux 
# (ici en x=-7.5 par exemple), et parvienne à trouver le 
# minimum global (ici x=-2 environ). D'autres n'y arrivent 
# pas, comme la descente de gradient.
# testons !
optimize.minimize(f, x0=-9)  # x0 est le point de départ
```

```python
# x: array([-7.0688912]) aie... c'est un minimum local ça...
optimize.minimize(f, x0=0)  # Testons en 0
```

```python
# Mieux ! Mais on voit bien que l'algo n'est pas tolérent aux minima locaux !
x0 = 1.5
res = optimize.minimize(f, x0=x0).x

plt.plot(x, f(x))
plt.scatter(x0, f(x0), s=100, c='g', marker='+')
plt.scatter(res, f(res), s=100, c='r')
```

```python
# testons avec une autre fonction :
f = lambda x: np.sin(x[0]) + np.cos(x[0] + x[1]) * np.cos(x[0])

# On créé les datas
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)

# On créé la grid
x, y = np.meshgrid(x, y)

plt.contour(x, y, f(np.array([x, y])), 20)

x0 = np.zeros((2, 1))
plt.scatter(x0[0], x0[1], marker='+', c='r', s=200)

res = optimize.minimize(f, x0=x0).x
plt.scatter(res[0], res[1], c='r', s=200)
```

```python
# Et si on veut le max ?
# facile, on inverse la fonction !

f = lambda x: -(np.sin(x[0]) + np.cos(x[0] + x[1]) * np.cos(x[0]))

# On créé les datas
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)

# On créé la grid
x, y = np.meshgrid(x, y)

plt.contour(x, y, f(np.array([x, y])), 20)

x0 = np.zeros((2, 1))
plt.scatter(x0[0], x0[1], marker='+', c='r', s=200)

res = optimize.minimize(f, x0=x0).x
plt.scatter(res[0], res[1], c='r', s=200)
```

## 4. Signal et FFT
signal.processing

signal.fftpack : transfo de fourrier

```python
from scipy import signal
```

```python
# on créé des data
x = np.linspace(0, 20, 100)
y = x + 4*np.sin(x) + np.random.randn(x.shape[0])

plt.plot(x, y)
```

```python
# on supprime l'évolution linéaire pour ne garder que le cycle
new_y = signal.detrend(y)

plt.plot(x,  y)
plt.plot(x,  new_y)
```

Transformation de Fourier : technique mathématique pour extraire et analyser les fréquences d'un signal périodique, afin d'obtenir un spectre de fréquences et d'amplitude.

Scipy propose un pack pour :

```python
from scipy import fftpack
```

```python
x = np.linspace(0, 30, 1000)
y = 3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x) + np.random.randn(x.shape[0])
```

```python
fourier = fftpack.fft(y)
freq = fftpack.fftfreq(y.size)

plt.plot(freq, fourier)
```

```python
power = np.abs(fourier)
plt.plot(np.abs(freq), power)
```

Maintenant on pourrait filtrer un signal :

```python
fourier[power<100] = 0

plt.plot(np.abs(freq), np.abs(fourier))
```

```python
# on réinverse la fonction de Fourier pour retrouver un signal:
filtered_signal = fftpack.ifft(fourier)

plt.figure(figsize=(16,9))
plt.plot(x, y, alpha=.5, c='g')
plt.plot(x, filtered_signal)
```

```python
plt.figure(figsize=(30, 5))

color = 0
isle_r = np.copy(img[:,:,color])

plt.subplot(1, 3, 1)
plt.imshow(isle_r, cmap="Reds")

plt.subplot(1, 3, 2)
plt.hist(isle_r.ravel(), bins=255);

isle_r = np.ones(img[:,:,color].shape)
isle_r[img[:,:,color] > 220] = 0
isle_r[img[:,:,color] < 70] = 0

plt.subplot(1, 3, 3)
plt.imshow(isle_r, cmap="Reds")
```

```python
plt.figure(figsize=(30, 5))

color = 1
isle_g = np.copy(img[:,:,color])

plt.subplot(1, 3, 1)
plt.imshow(isle_g, cmap="Greens")

plt.subplot(1, 3, 2)
plt.hist(isle_g.ravel(), bins=255);

isle_g = np.ones(img[:,:,color].shape)
isle_g[img[:,:,color] > 230] = 0
isle_g[img[:,:,color] < 50] = 0

plt.subplot(1, 3, 3)
plt.imshow(isle_g, cmap="Greens")
```

```python
# On va tenter d'extraire les iles, utilisons les bleu
plt.figure(figsize=(30, 5))

color = 2
isle_b = np.copy(img[:,:,color])

plt.subplot(1, 3, 1)
plt.imshow(isle_b, cmap="Blues")

plt.subplot(1, 3, 2)
plt.hist(isle_b.ravel(), bins=255);

isle_b = np.zeros(img[:,:,color].shape)
isle_b[img[:,:,color] < 40] = 1

plt.subplot(1, 3, 3)
plt.imshow(isle_b, cmap="Blues")
```

```python
plt.figure(figsize=(30, 10))

plt.subplot(1, 3, 1)
plt.imshow(img)


isle_mask = isle_b + isle_g + isle_r
isle_mask[isle_mask < 2] = 0
isle_mask[isle_mask > 1] = 1
plt.subplot(1, 3, 2)
plt.imshow(isle_mask, cmap="gray")


img_clean = np.copy(img)
img_clean[:,:,0] = img[:,:,0] * isle_mask
img_clean[:,:,1] = img[:,:,1] * isle_mask
img_clean[:,:,2] = img[:,:,2] * isle_mask

plt.subplot(1, 3, 3)
plt.imshow(img_clean)
```

```python
plt.figure(figsize=(30, 10))

plt.subplot(1, 3, 1)
plt.imshow(img)


img_clean_gray = ndimage.binary_opening(img_clean.dot([0.07, 0.72, 0.21]))
plt.subplot(1, 3, 2)
plt.imshow(img_clean_gray, cmap="gray")


img_clean = np.copy(img)
img_clean[:,:,0] = img[:,:,0] * img_clean_gray
img_clean[:,:,1] = img[:,:,1] * img_clean_gray
img_clean[:,:,2] = img[:,:,2] * img_clean_gray

plt.subplot(1, 3, 3)
plt.imshow(img_clean)
```

```python
# Tentons de mettre des labels
label_image, n_labels = ndimage.label(img_clean_gray)

print(f'{n_labels} labels créés')
plt.imshow(label_image)
```

```python
sizes = ndimage.sum(img_clean_gray, label_image, range(n_labels))
plt.scatter(range(n_labels), sizes)
```
